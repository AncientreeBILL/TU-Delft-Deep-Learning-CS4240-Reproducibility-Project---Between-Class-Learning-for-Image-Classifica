```
path = '/content/drive/My Drive/DL'
```

```
standa = np.loadtxt(path + '/cnn_stan.txt')
bc = np.loadtxt(path + '/cnn_bc.txt')

bcp = []
with open(path + '/cnn_bc_plus.txt') as p:
  lines = p.readlines()
  for line in lines:
    bcp.append(line.split())

bc_p = []
for i in range(0, len(bcp)):
  bc_p = np.append(bc_p, float("".join(bcp[i])))
```

```
x = range(0, 250)

plt.plot(x, standa,  color='red', label='Standard')
plt.plot(x, bc,  color='skyblue', label='BC')
plt.plot(x, bc_p, color='blue', label='BC+')
plt.legend() 

plt.xlabel('epoch')
plt.ylabel('top1 error rate')
plt.show()
```
